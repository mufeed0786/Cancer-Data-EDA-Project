# **Exploratory Data Analysis**

1. Genetic_Risk VS Target_Severity_Score

```python
cancer_patients.groupby('Genetic_Risk')['Target_Severity_Score'].mean()
```

Observation --
“Severity score increases with genetic risk, indicating a positive relationship, although the increase is moderate.”

2. Smoking vs Target_Severity_Score

```python
cancer_patients.groupby('Smoking')['Target_Severity_Score'].mean()
```

“Severity score increases with higher smoking levels, indicating a positive relationship between smoking and cancer severity"

3. Air_Pollution vs Target_severity_score

```python
cancer_patients.groupby('Air_Pollution')['Target_Severity_Score'].mean()
```

Observation:-
“Severity score increases with higher air pollution levels, indicating a moderate positive relationship.”

4. Alcohol_Use VS Target_severity_score

```python
cancer_patients.groupby('Alcohol_Use')['Target_Severity_Score'].mean()
```

Observation:-
“Severity score increases with higher alcohol consumption, indicating a moderate positive relationship between alcohol use and cancer severity.”

5. Obesity_Level vs Target_Severity_Score

```python
cancer_patients.groupby('Obesity_Level')['Target_Severity_Score'].mean()
```

Observation:-
“Severity score increases with higher obesity levels, indicating a moderate positive relationship between obesity and cancer severity.”

6. Cancer_Type vs Target_Severity_Score

```python
cancer_patients.groupby('Cancer_Type')['Target_Severity_Score'].mean()
```

Observation:-

7. Cancer_Stage vs Target_Severity_Score

“Severity scores are very similar across different cancer stages, with only minor variations, suggesting that no cancer stage is significantly more severe than other.”

# **Graphs**

1.Smoking vs severity

```python
cancer_patients.groupby('Smoking')['Target_Severity_Score'].mean().plot()
plt.title("Smoking vs Severity")
plt.xlabel("Smoking Level")
plt.ylabel("Average Severity Score")
plt.show()
```

```python
cancer_patients.groupby('Genetic_Risk')['Target_Severity_Score'].mean().plot()
plt.title("Genetic Risk vs Severity")
plt.xlabel("Genetic Risk")
plt.ylabel("Average Severity Score")
plt.show()
```

# Correlation

```python
cancer_patients.corr(numeric_only=True)
```

Insight--              
Smoking and genetic risk show the strongest positive correlation with severity score. Air pollution and alcohol use show moderate relationships, while obesity has a weaker effect. Treatment cost shows a negative correlation, indicating that higher costs are associated with lower severity.

# **Final Conclusion**


This project analyzed a global cancer dataset of 50,000 patients to understand the impact of various risk factors on cancer severity. The analysis shows that smoking and genetic risk have the strongest positive relationship with severity. Air pollution, alcohol use, and obesity exhibit moderate effects. In contrast, treatment cost shows a negative relationship with severity. Additionally, cancer type and stage do not show significant variation in severity within this dataset. Overall, lifestyle and genetic factors appear to play a more important role in determining cancer severity.
